# Create data for survival analysis

## Read data files

### Read file containing journeys

```python
import os

from pathlib import Path

import pandas as pd


journeysDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'data', 'wb_365_wa_3', 'data_matrix_mortality.csv'))
journeysDf
```

### Read file containing data

```python
import os

from pathlib import Path

import pandas as pd


dataDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240531', 'data', 'data_matrix.csv'))
dataDf
```

## Merge data files

```python
mergedDf = journeysDf[['person_id', 'visit_occurrence_id', 'JOURNEY_ID']].drop_duplicates().merge(
    dataDf[['person_id', 'visit_occurrence_id', 'admittime_adm', 'deathtime_adm']].drop_duplicates(),
    how='inner',
    on=['person_id', 'visit_occurrence_id']
).drop(
    columns=['visit_occurrence_id']
).drop_duplicates()
mergedDf
```

## Save journey time information for survival analysis

```python
import os

from pathlib import Path


mergedDf.to_csv(Path(os.environ['EHR_DATA_BASE'], 'data', 'full_cohort', 'journry_times.csv'), index=False)
```
