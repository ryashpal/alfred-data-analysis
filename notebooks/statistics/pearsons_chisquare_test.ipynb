## Merge all the information

```python
mergedDf = mlstDf[['tube_code', 'mlst_type']].merge(
    patientTubeidMappingDf[['tube_code', 'PATIENT_ID', 'date_of_collection']].drop_duplicates(),
    how='inner',
    on=['tube_code'],
).merge(
    admissionsDf[['PATIENT_ID', 'DATEOFDEATH_DATETIME']].drop_duplicates(),
    how='inner',
    on=['PATIENT_ID']
)
mergedDf
```

```python
mergedDf['DATEOFDEATH_DATETIME'] = pd.to_datetime(mergedDf['DATEOFDEATH_DATETIME'], format='%Y-%m-%d %H:%M:%S.%f')
mergedDf['date_of_collection'] = pd.to_datetime(mergedDf['date_of_collection'], format='%Y-%m-%d')
mergedDf
```

```python
mergedDf['death_30_days'] = (((mergedDf.DATEOFDEATH_DATETIME - mergedDf.date_of_collection).dt.days >= 0) & ((mergedDf.DATEOFDEATH_DATETIME - mergedDf.date_of_collection).dt.days < 30))
mergedDf['death_30_days'] = mergedDf['death_30_days'].apply(lambda x: 1 if x else 0)
mergedDf = mergedDf.drop(columns=['date_of_collection', 'DATEOFDEATH_DATETIME'])
mergedDf
```

```python
aggregatedDf = mergedDf.groupby(['mlst_type'])\
    .agg({'death_30_days': ['sum', 'count']})\
        .reset_index()\
            .pipe(lambda df: df.set_axis(
                ['_'.join(col) if col[1] else col[0] for col in df.columns],
                axis=1
            ))
aggregatedDf
```

```python
aggregatedDf['expected'] = aggregatedDf.death_30_days_count*aggregatedDf.death_30_days_sum.sum()/aggregatedDf.death_30_days_count.sum()
aggregatedDf
```

```python
aggregatedDf['difference'] = (aggregatedDf.death_30_days_sum - aggregatedDf.expected)/aggregatedDf.expected * 100
aggregatedDf.sort_values(by=['difference'])
```

```python
aggregatedDf['percentage'] = aggregatedDf.death_30_days_sum/aggregatedDf.death_30_days_count * 100
aggregatedDf.sort_values(by=['difference'])
```
