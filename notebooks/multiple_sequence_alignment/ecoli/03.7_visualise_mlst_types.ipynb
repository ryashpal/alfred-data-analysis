# Visualise location information

```r
options(repr.plot.width = 16, repr.plot.height = 12)

library(ape)
library(ggtree)
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)

tree <- read.tree(
  "~/workspace/genome_data/annotations/e_coli/gene_sequences/acrF_tree.nwk"
)

mlst_df <- read_csv(
  "~/workspace/genome_data/annotations/e_coli/mlst/mlst_ecoli.csv"
)

df <- data.frame(tube_code = tree$tip.label)

heatmap_data <- df %>%
  left_join(mlst_df, by = "tube_code")

circ <- ggtree(tree, layout = "circular", branch.length = "none")

rownames(heatmap_data) <- heatmap_data$tube_code

p <- gheatmap(
  circ,
  heatmap_data %>% select(mlst_type),
  offset = .8,
  width = .2,
  colnames = FALSE
)

p
```

```r
options(repr.plot.width = 16, repr.plot.height = 12)

library(ape)
library(ggtree)
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)
library(ggnewscale)

tree <- read.tree(
  "~/workspace/genome_data/annotations/e_coli/gene_sequences/acrF_tree.nwk"
)

mlst_df <- read_csv(
  "~/workspace/genome_data/annotations/e_coli/mlst/mlst_ecoli.csv"
)

df <- data.frame(tube_code = tree$tip.label)

heatmap_data <- df %>%
  left_join(mlst_df, by = "tube_code")

circ <- ggtree(tree, layout = "circular", branch.length = "none")

rownames(heatmap_data) <- heatmap_data$tube_code

p <- gheatmap(
  circ,
  heatmap_data %>% select(mlst_type),
  offset = .8,
  width = .2,
  colnames = FALSE
) +
  new_scale_fill()

p <- gheatmap(
  p,
  heatmap_data %>% select(adk),
  offset = 10,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(fumC),
  offset = 20,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(gyrB),
  offset = 30,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(icd),
  offset = 40,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(mdh),
  offset = 50,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(purA),
  offset = 60,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(recA),
  offset = 70,
  width = .2,
  colnames = FALSE
)

p
```
