## Format output

```python
mlstOutputDf
```

```python
mlstOutputDf['tube_code'] = mlstOutputDf.file_name.str[47:55]
mlstOutputDf['arcC'] = mlstOutputDf.mlst_type_1.str.strip('arcC').str.strip('(').str.strip(')')
mlstOutputDf['aroE'] = mlstOutputDf.mlst_type_2.str.strip('aroE').str.strip('(').str.strip(')')
mlstOutputDf['glpF'] = mlstOutputDf.mlst_type_3.str.strip('glpF').str.strip('(').str.strip(')')
mlstOutputDf['gmk'] = mlstOutputDf.mlst_type_4.str.strip('gmk').str.strip('(').str.strip(')')
mlstOutputDf['pta'] = mlstOutputDf.mlst_type_5.str.strip('pta').str.strip('(').str.strip(')')
mlstOutputDf['tpi'] = mlstOutputDf.mlst_type_6.str.strip('tpi').str.strip('(').str.strip(')')
mlstOutputDf['yqiL'] = mlstOutputDf.mlst_type_7.str.strip('yqiL').str.strip('(').str.strip(')')
mlstOutputDf = mlstOutputDf.drop(columns=['file_name', 'mlst_type_1', 'mlst_type_2', 'mlst_type_3', 'mlst_type_4', 'mlst_type_5', 'mlst_type_6', 'mlst_type_7'])
mlstOutputDf
```

```python
topMlstTypesList = mlstOutputDf[mlstOutputDf.mlst_type != '-'].mlst_type.value_counts().reset_index().sort_values(by=['count'], ascending=False).mlst_type[:10]
topMlstTypesList
```

```python
mlstOutputDf = mlstOutputDf[mlstOutputDf.mlst_type.isin(topMlstTypesList)]
mlstOutputDf
```

## Save formatted mlst output

```python
import os
from pathlib import Path

import pandas as pd


mlstOutputFormattedDir = Path(os.environ['GENOMICS_DATA_BASE'], 'annotations', 's_aureus', 'mlst')
mlstOutputFormattedFile = Path(mlstOutputFormattedDir, 'mlst_saur.csv')


mlstOutputDf.to_csv(mlstOutputFormattedFile, index=False)
```
