# Visualise location information

```r
options(repr.plot.width = 16, repr.plot.height = 12)

library(ape)
library(ggtree)
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)

tree <- read.tree(
  "~/workspace/genome_data/annotations/s_aureus/gene_sequences/essC/full/essC_tree.nwk"
)

mlst_df <- read_csv(
  "~/workspace/genome_data/annotations/s_aureus/mlst/mlst_saur.csv"
)

df <- data.frame(tube_code = tree$tip.label)

heatmap_data <- df %>%
  left_join(mlst_df, by = "tube_code")

circ <- ggtree(tree, layout = "circular", branch.length = "none")

rownames(heatmap_data) <- heatmap_data$tube_code

p <- gheatmap(
  circ,
  heatmap_data %>% select(mlst_type),
  offset = .8,
  width = .2,
  colnames = FALSE
)

p
```

```r
options(repr.plot.width = 16, repr.plot.height = 12)

library(ape)
library(ggtree)
library(ggplot2)
library(dplyr)
library(readr)
library(tidyr)
library(ggnewscale)

tree <- read.tree(
  "~/workspace/genome_data/annotations/s_aureus/gene_sequences/essC/full/essC_tree.nwk"
)

mlst_df <- read_csv(
  "~/workspace/genome_data/annotations/s_aureus/mlst/mlst_saur.csv"
)

df <- data.frame(tube_code = tree$tip.label)

heatmap_data <- df %>%
  left_join(mlst_df, by = "tube_code")

circ <- ggtree(tree, layout = "circular", branch.length = "none")

rownames(heatmap_data) <- heatmap_data$tube_code

p <- gheatmap(
  circ,
  heatmap_data %>% select(mlst_type),
  offset = .8,
  width = .2,
  colnames = FALSE
) +
  new_scale_fill()

p <- gheatmap(
  p,
  heatmap_data %>% select(arcC),
  offset = 10,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(aroE),
  offset = 20,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(glpF),
  offset = 30,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(gmk),
  offset = 40,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(pta),
  offset = 50,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(tpi),
  offset = 60,
  width = .2,
  colnames = FALSE
)

p <- gheatmap(
  p,
  heatmap_data %>% select(yqiL),
  offset = 70,
  width = .2,
  colnames = FALSE
)

p
```
