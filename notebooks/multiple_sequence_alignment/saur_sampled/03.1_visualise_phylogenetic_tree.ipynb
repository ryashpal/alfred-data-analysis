# Visualise phylogenetic tree

```r
options(repr.plot.width = 10, repr.plot.height = 15)


library(ape)
library(ggtree)
library(ggplot2)

tree <- read.tree(
  "~/workspace/genome_data/annotations/s_aureus/gene_sequences/essC/sample/tree.nwk"
)

p <- ggtree(tree, branch.length = "none")

p
```
