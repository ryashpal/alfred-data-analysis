# Obtain cohort sizes

## Mortality

```python
import os

import pandas as pd

from pathlib import Path


testDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'data', 'wb_365_wa_3', 'splits_v1', 'labels', 'mortality', 'mortality_30_day_normal_test.csv'), sep='\t')
trainDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'data', 'wb_365_wa_3', 'splits_v1', 'labels', 'mortality', 'mortality_30_day_normal_train.csv'), sep='\t')
validateDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'data', 'wb_365_wa_3', 'splits_v1', 'labels', 'mortality', 'mortality_30_day_normal_validate.csv'), sep='\t')

fullDf = pd.concat([testDf, trainDf, validateDf], ignore_index=True)
fullDf
```

```python
fullDf[['person_id', 'JOURNEY_ID', 'death_30_day']].drop_duplicates().death_30_day.value_counts()/fullDf[['person_id', 'JOURNEY_ID']].drop_duplicates().shape[0]*100
```
