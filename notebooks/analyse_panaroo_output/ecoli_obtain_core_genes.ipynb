# Obtain core genes

```python
import os
from pathlib import Path

import pandas as pd


genePresenceAbsenceFile = Path(os.environ['GENOMICS_DATA_BASE'], 'annotations', 'e_coli_panaroo_strict', 'gene_presence_absence.csv')
genePresenceAbsenceDf = pd.read_csv(genePresenceAbsenceFile)
genePresenceAbsenceDf
```

```python
genePresenceAbsenceDf = genePresenceAbsenceDf.drop(columns=['Non-unique Gene name', 'Annotation'])
genePresenceAbsenceDf = genePresenceAbsenceDf.set_index('Gene')
genePresenceAbsenceTDf = genePresenceAbsenceDf.T
genePresenceAbsenceTDf
```

```python
geneCountsDf = genePresenceAbsenceTDf.notna().sum().reset_index().rename(columns={0: 'Occurrences'})
geneCountsDf
```

```python
geneCountsDf['Occurrences'].value_counts().reset_index().sort_values(by=['Occurrences'], ascending=False)[:20]
```

```python
geneCountsDf['Occurrences'].value_counts().reset_index().sort_values(by=['Occurrences'], ascending=False)[-20:]
```

```python
geneOccurancesDf = geneCountsDf['Occurrences'].value_counts().reset_index().sort_values(by=['Occurrences'], ascending=True)
geneOccurancesDf['Gene Proportion'] = geneOccurancesDf['count'].cumsum()/geneOccurancesDf['count'].sum()
thresholdGeneOccurance = geneOccurancesDf[geneOccurancesDf['Gene Proportion'] > 0.95].sort_values(by='Gene Proportion', ascending=True).Occurrences.values[0]
thresholdGeneOccurance
```

```python
coreGenesDf = geneCountsDf[geneCountsDf['Occurrences'] >= thresholdGeneOccurance]
coreGenesDf
```

```python
import os
from pathlib import Path


coreGenesDf.to_csv(Path(os.environ['GENOMICS_DATA_BASE'], 'annotations', 'e_coli_panaroo_strict', 'core_genes.csv'), index=False)
```
