## Read data matrix

```python
import os

from pathlib import Path

import pandas as pd


dmDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'data', 'wb_365_wa_3', 'data_matrix_mortality.csv'))
dmDf
```

## Merge data

```python
mergedDf = demoDf.merge(
    dmDf,
    how='inner',
    left_on=['PATIENT_ID', 'EPISODE_ID'],
    right_on=['person_id', 'visit_occurrence_id'],
).drop(
    columns=['person_id', 'visit_occurrence_id']
)
mergedDf
```

```python
countryCountsDf = mergedDf[['PATIENT_ID', 'COUNTRY_OF_BIRTH']].drop_duplicates().groupby(by=['COUNTRY_OF_BIRTH']).agg(func='count').reset_index().rename(columns={'PATIENT_ID': 'Count'}).sort_values(by=['Count'], ascending=False)
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'England'), 'COUNTRY_OF_BIRTH'] = 'United Kingdom'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'United Kingdom, Channel Islands & Isle of Man'), 'COUNTRY_OF_BIRTH'] = 'United Kingdom'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Scotland'), 'COUNTRY_OF_BIRTH'] = 'United Kingdom'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Wales'), 'COUNTRY_OF_BIRTH'] = 'United Kingdom'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'United States of America'), 'COUNTRY_OF_BIRTH'] = 'United States'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Australian External Territories'), 'COUNTRY_OF_BIRTH'] = 'Australia'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Australian Antarctic Territory'), 'COUNTRY_OF_BIRTH'] = 'Australia'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'China (Excludes SARS & Taiwan)'), 'COUNTRY_OF_BIRTH'] = 'China'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Hong Kong(Special Administrative Region of China)'), 'COUNTRY_OF_BIRTH'] = 'China'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Chinese Asia (includes Mongolia)'), 'COUNTRY_OF_BIRTH'] = 'China'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Macau (Special Administration Region of China)'), 'COUNTRY_OF_BIRTH'] = 'China'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Japan and the Koreas, nfd'), 'COUNTRY_OF_BIRTH'] = 'Japan'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Russian Federation'), 'COUNTRY_OF_BIRTH'] = 'Russia'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'East Timor'), 'COUNTRY_OF_BIRTH'] = 'Timor-Leste'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Burma (Myanmar)'), 'COUNTRY_OF_BIRTH'] = 'Myanmar'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Aland Islands'), 'COUNTRY_OF_BIRTH'] = 'Finland'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Former Yugoslav Republic of Macedonia'), 'COUNTRY_OF_BIRTH'] = 'Macedonia'
countryCountsDf.loc[(countryCountsDf.COUNTRY_OF_BIRTH == 'Bosnia-Herzegovina'), 'COUNTRY_OF_BIRTH'] = 'Bosnia And Herzegovina'


countryCountsDf
```

## Read country to region mapping data

```python
import os

from pathlib import Path

import pandas as pd


mapDf = pd.read_csv(Path(os.environ['EHR_DATA_BASE'], 'blood_pos_cohort_20240614', 'etc', 'continents2.csv'))
mapDf
```

```python
mappedDf = countryCountsDf.merge(
    mapDf,
    how='left',
    left_on=['COUNTRY_OF_BIRTH'],
    right_on=['name']
)
mappedDf
```

## Plot

### Plot regions

```python
regionDf = mappedDf.groupby(by=['region']).agg(func={'Count': 'sum'}).reset_index().sort_values(by=['Count'])
regionDf
```

```python
import matplotlib.pyplot as plt 
import seaborn 

fig, ax = plt.subplots(figsize=(8, 8))

# define Seaborn color palette to use 
palette_color = seaborn.color_palette("pastel")

# plotting data on chart 
plt.pie(regionDf.Count, labels=regionDf.region, colors=palette_color, autopct='%.0f%%') 

ax.set_title('Region wise patient country of birth')

# displaying chart 
plt.show()
```

### Plot sub-regions

```python
subregionDf = mappedDf.groupby(by=['sub-region']).agg(func={'Count': 'sum'}).reset_index().sort_values(by=['Count'])
subregionDf.loc[(subregionDf['sub-region'] == 'Melanesia'), 'sub-region'] = 'Pacific Islands'
subregionDf.loc[(subregionDf['sub-region'] == 'Polynesia'), 'sub-region'] = 'Pacific Islands'
subregionDf.loc[(subregionDf['sub-region'] == 'Central Asia'), 'sub-region'] = 'Eastern Asia'
subregionDf.loc[(subregionDf['sub-region'] == 'Central Asia'), 'sub-region'] = 'Eastern Asia'
subregionDf.loc[(subregionDf['sub-region'] == 'Northern Africa'), 'sub-region'] = 'Northern & Sub-Saharan Africa'
subregionDf.loc[(subregionDf['sub-region'] == 'Sub-Saharan Africa'), 'sub-region'] = 'Northern & Sub-Saharan Africa'
subregionDf = subregionDf.groupby(by=['sub-region']).agg(func={'Count': 'sum'}).reset_index().sort_values(by=['Count'])
subregionDf
```

```python
import matplotlib.pyplot as plt 
import seaborn 


fig, ax = plt.subplots(figsize=(8, 8))

# define Seaborn color palette to use 
palette_color = seaborn.color_palette("pastel")

# plotting data on chart 
plt.pie(subregionDf.Count, labels=subregionDf['sub-region'], colors=palette_color, autopct='%.0f%%') 

ax.set_title('Sub-region wise patient country of birth')
# displaying chart 
plt.show()
```
