# Visualising enriched GO pathways

## ECOLI

### Read Data

```python
ecoliDataString = """1	transporter activity (GO:0005215)	38	11.40%	17.80%	Molecular Function
2	electron transfer activity (GO:0009055)	1	0.30%	0.50%	Molecular Function
3	translation regulator activity (GO:0045182)	2	0.60%	0.90%	Molecular Function
4	catalytic activity (GO:0003824)	108	32.30%	50.70%	Molecular Function
5	molecular function regulator activity (GO:0098772)	2	0.60%	0.90%	Molecular Function
6	ATP-dependent activity (GO:0140657)	19	5.70%	8.90%	Molecular Function
7	molecular transducer activity (GO:0060089)	1	0.30%	0.50%	Molecular Function
8	antioxidant activity (GO:0016209)	1	0.30%	0.50%	Molecular Function
9	binding (GO:0005488)	41	12.30%	19.20%	Molecular Function
1	detoxification (GO:0098754)	1	0.30%	0.40%	Biological Process
2	response to stimulus (GO:0050896)	16	4.80%	6.70%	Biological Process
3	cellular process (GO:0009987)	107	32.00%	44.60%	Biological Process
4	metabolic process (GO:0008152)	82	24.60%	34.20%	Biological Process
5	homeostatic process (GO:0042592)	2	0.60%	0.80%	Biological Process
6	biological regulation (GO:0065007)	7	2.10%	2.90%	Biological Process
7	localization (GO:0051179)	25	7.50%	10.40%	Biological Process
1	cellular anatomical entity (GO:0110165)	123	36.80%	88.50%	Cellular Component
2	protein-containing complex (GO:0032991)	16	4.80%	11.50%	Cellular Component"""
ecoliData = list(map(lambda x: x.split('\t'), ecoliDataString.split('\n')))
# .apply()
ecoliData
```

```python
import pandas as pd


ecoliDf = pd.DataFrame(ecoliData, columns=['ix', 'GO term', 'Count', 'Percentage 1', 'Percentage 2', 'Ontology'])
ecoliDf
```

```python
molecularDf = ecoliDf[ecoliDf.Ontology == 'Molecular Function']
biologicalDf = ecoliDf[ecoliDf.Ontology == 'Biological Process']
cellularDf = ecoliDf[ecoliDf.Ontology == 'Cellular Component']
pathwayDf = ecoliDf[ecoliDf.Ontology == 'Pathway']
```

```python
from matplotlib import pyplot as plt
import seaborn as sns


# f, axes = plt.figure(figsize=(12, 8)).subplots(2, 2)
f, axes = plt.subplots(2, 2, figsize=(16, 8))


colors = sns.color_palette('pastel')

#create pie chart
axes[0, 0].pie(x=molecularDf['Count'], colors = colors, autopct='%.0f%%')
axes[0, 0].legend(labels=molecularDf['GO term'], loc='center right', bbox_to_anchor=(1.92, .5))
axes[0, 0].set_title('(A) Molecular Functions')

axes[1, 0].pie(x=biologicalDf['Count'], colors = colors, autopct='%.0f%%')
axes[1, 0].legend(labels=biologicalDf['GO term'], loc='center right', bbox_to_anchor=(1.9, .5))
axes[1, 0].set_title('(B) Biological Processes')

axes[0, 1].pie(x=cellularDf['Count'], colors = colors, autopct='%.0f%%')
axes[0, 1].legend(labels=cellularDf['GO term'], loc='center right', bbox_to_anchor=(2.1, .5))
axes[0, 1].set_title('(C) Cellular Components')

axes[1, 1].pie(x=pathwayDf['Count'], colors = colors, autopct='%.0f%%')
axes[1, 1].legend(labels=pathwayDf['GO term'], loc='center right', bbox_to_anchor=(2.1, .5))
axes[1, 1].set_title('(D) Pathways')

plt.suptitle('Gene Ontology Enrichment Analysis Results')
plt.show()
```
